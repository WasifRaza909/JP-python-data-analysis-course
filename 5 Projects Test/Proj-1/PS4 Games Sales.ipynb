```python
sorted_data = df.sort_values(by="Global", ascending=False)
```

```python
top_10 = sorted_data.head(5)
top_10
```

```python
plt.bar(top_10["Game"], top_10["Global"])
plt.xlabel("Games")
plt.ylabel("Sales")
plt.xticks(rotation = 45, fontsize=8)
plt.title("Top 5 Games by Global Sales")
plt.show()
```

```python
region_sales = df.groupby('Genre')[['North America', 'Europe', 'Japan', 'Rest of World']].sum()
region_sales.plot(kind="bar", figsize=(12,6))
plt.ylabel("Sales")
plt.title("Sales Comparison by Region & Genres",fontsize=18)
```

```python
region_sales = df.groupby('Publisher')[['Global']].sum().reset_index().sort_values(by='Global', ascending=False).head(5)
region_sales.plot(kind="bar", figsize=(12,6))
plt.ylabel("Sales")
plt.title("Top 5 Publisher Sales Globally",fontsize=18)
plt.xticks(ticks=range(len(region_sales)), labels=region_sales['Publisher'], rotation = 45, fontsize=8)
plt.show()
region_sales['Global'].iloc[0]
```

### The above chart shows that the top Global Seller is Activision with sale of 72.44

```python
years = df["Year"]

df_years =  pd.DataFrame(years.value_counts().reset_index()) 

plt.bar(df_years["Year"], df_years["count"])
plt.xlabel("Years")
plt.ylabel("Releases Counts")
plt.xticks(rotation = 45, fontsize=8)
plt.title("Releases Counts By Years")
plt.show()
```

### The above chart indicates most games are released in 2017

```python
global_sold_genre = df.groupby("Genre")['Global'].sum().reset_index().head(5)
global_sold_genre.plot(kind="bar", figsize=(12,6))
plt.ylabel("Sales")
plt.title("Top 5 Genre Sales Globally",fontsize=18)
plt.xticks(ticks=range(len(global_sold_genre)), labels=global_sold_genre['Genre'], rotation = 45, fontsize=8)
plt.show()
global_sold_genre['Global'].iloc[0]
```

### The above chart shows that the top Globally Sold Genre is Action with sale of 136.85

```python
top_5_games_in_japan = df.groupby('Game')['Japan'].sum().reset_index().sort_values(by='Japan', ascending=False).head(5)
top_5_games_in_japan.plot(kind="bar", figsize=(12,6))
                                              
plt.ylabel("Sales")
plt.title("Top 5 Best selling games in Japan",fontsize=18)
plt.xticks(ticks=range(len(top_5_games_in_japan)), labels=top_5_games_in_japan['Game'])

top_played_game_in_japan = top_5_games_in_japan["Game"].iloc[0]
game_details = df[df["Game"] == top_played_game_in_japan]
game_details_group = game_details.groupby("Game")[['North America', 'Europe', 'Japan', 'Rest of World']].sum()
game_details_group.plot(kind="bar", figsize=(12,6))
plt.xlabel("")
plt.title(f'Sales of "{top_played_game_in_japan}" in different regions',fontsize=18) 
plt.xticks(rotation = 45, fontsize=8)
plt.show()
```

### The above charts shows that the most played game in Japan is "Monster Hunter: World"

```python
correlation_matrix = df[['North America', 'Europe']].corr()

# Plot the heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap: North America vs Europe Sales')
plt.show()
```

## The above heatmap shows there is a good coorelation between Europe and North America

```python
df_action = df[df["Genre"] == "Action"].groupby("Year")["Global"].sum()

df_action.plot(kind="line", figsize=(12,6) , label="Action Games")
plt.title("Yearly Trend Analysis Action Games", fontsize= 18)
plt.ylabel("Sales")

df_shooter = df[df["Genre"] == "Shooter"].groupby("Year")["Global"].sum()

df_shooter.plot(kind="line", figsize=(12,6), label="Shooter Games")
plt.title("Yearly Trend Analysis Comparison Shooter and Action Games", fontsize= 18)
plt.ylabel("Sales")
plt.legend()

plt.show()
```
